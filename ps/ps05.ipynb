3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
boxcar = np.ones(11)
boxcar /= sum(boxcar)
boxcar = thinkdsp.Spectrum(boxcar, 10)
boxcar = boxcar.make_wave().make_spectrum()
boxcar.plot()
```

```python
thinkplot.plot(boxcar.fs, boxcar.angles())
```

```python
import scipy.signal
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
gaussian = thinkdsp.Spectrum(gaussian, 10)
gaussian.plot()
```

```python
thinkplot.plot(gaussian.fs, gaussian.angles())
```

There isn't a phase shift with either boxcar or gaussian windows. They only attenuate or amplify the signal, no phase shifts occur.

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
w_dw_s = np.array([-0.5,2,-3.0/2]) # second order approximation of first derivative
size = 200
w_dw_s = np.hstack((w_dw_s, np.zeros(size)))
s_dw_s = thinkdsp.Wave(w_dw_s, 1).make_spectrum()
s_dw_s.plot()
```

This looks like a high pass filter. The dropoff is not as great as the discontinuous "perfect" filter from above.

```python
s_dw_s.make_wave().plot()
```

```python
w_ddw_f = np.array([1.0, -2.0, 1.0]) # approximation of second derivative
w_ddw_f = np.hstack((w_ddw_f, np.zeros(size)))
s_ddw_f = thinkdsp.Wave(w_ddw_f, 1).make_spectrum()
s_ddw_f.plot()
```

I think the nth derivative approximation makes better and better high pass filters because the dropoff is even steeper. In this case, this is the second derivative, which is much better than the first derivative above.

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
psig = thinkdsp.ParabolicSignal(freq=4)
pwave = psig.make_wave(start=0, duration=2.0)
pwave.plot()
```

```python
pderiv = pwave.make_spectrum()
pderiv.differentiate()
pderiv.make_wave().plot()
```

I knew it! The saw tooth wave is the derivative of the parabolic wave! This was the solution to a question in a previous problem set. It also makes sense that the derivative of a parabola is linear, therefore the derivative of a parabolic signal is a linear signal (saw tooth).
