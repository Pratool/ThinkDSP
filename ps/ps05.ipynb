2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

```python
quality = 10000
lp_hs = np.hstack((np.zeros(quality)+1, np.zeros(quality)))
lp_filter = thinkdsp.Spectrum(lp_hs, 3000)
lp_filter.plot()
```

```python
lp_filter.make_wave().plot()
```

I found this really intriguing. Convolving this sort of wave in the time domain with another wave will apply a "perfect" low-pass filter. The low-pass filter's cutoff frequency is a quarter of the sample rate. I'm interested in what happens at the end. Perhaps a perfect (discrete) low-pass filter doesn't have any kind of ramp up and just has one point that is higher than the rest at the very end. Also, after closer inspection, I think the beginning also has a high amplitude, but is much harder to see.

```python
seg = lp_filter.make_wave().segment(start=13.0, duration=0.5)
seg.plot()
```

So there is a ramp up! I wonder at what rate it ramps up at. Let's take a closer look.

```python
seg2 = lp_filter.make_wave().segment(start=0, duration=0.5)
seg2.plot()
```

There is also a ramp down at the beginning. I get the feeling that the two ends of the wave are symmetrical.

```python
seg_seg = seg.segment(start=0.3, duration=0.05)
seg_seg.plot()
```

I can't zoom more than this because I can already see the discrete points. I see a sinusoidal-like wave (it looks more like a triangle wave because of aliasing... I think) that increases amplitude smoothly and in a concave up manner. I don't know what defines that increase, but I'd reckon it's either exponential or quadratic.

```python
hp_hs = np.hstack((np.zeros(quality), np.zeros(quality)+1))
hp_filter = thinkdsp.Spectrum(hp_hs, 3000)
hp_filter.plot()
```

```python
hp_filter.make_wave().plot()
```

```python
seg_hp = hp_filter.make_wave().segment(start=13.0, duration=0.5)
seg_hp.plot()
```

For the high-pass filter, the ending point is negative instead of positive (compared to the low-pass filter). This is indicative of a phase shift because the amplitude stays the same.

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
boxcar = np.ones(11)
boxcar /= sum(boxcar)
boxcar = thinkdsp.Spectrum(boxcar, 10)
boxcar = boxcar.make_wave().make_spectrum()
boxcar.plot()
```

```python
thinkplot.plot(boxcar.fs, boxcar.angles())
```

```python
import scipy.signal
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
gaussian = thinkdsp.Spectrum(gaussian, 10)
gaussian = gaussian.make_wave().make_spectrum()
gaussian.plot()
```

```python
thinkplot.plot(gaussian.fs, gaussian.angles())
```

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?
