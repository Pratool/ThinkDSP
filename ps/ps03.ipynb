1) At http://www.coindesk.com you can download the daily price of a BitCoin as a CSV file. Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
def read_csv(file_name, nrows):
    csv_reader = csv.reader(open(file_name));
    bitcoin_vec = []
    date_vec = []
    i = 0 #iterator to prevent the last row from being read

    # read each line
    for line in csv_reader: #each line is a list (NOTE: reads as string)
        if i <= nrows:
            bitcoin_vec.append(line[1])
            date_vec.append(line[0])
        i=i+1
    
    # Delete first entry in vectors
    del bitcoin_vec[0]
    del date_vec[0]

    # Convert strings to floats
    for i in range(len(bitcoin_vec)):
        bitcoin_vec[i] = float(bitcoin_vec[i])

    # Convert list to numpy array
    bitcoin_vec = np.array(bitcoin_vec)
    
    return [bitcoin_vec, date_vec]

data,_ = read_csv('bitcoin-data.csv', 2449)
wave = thinkdsp.Wave(data, framerate=1)
wave.plot()
```

```python
wave.make_spectrum().plot_power(low=1, linewidth=1)
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 legend=False)
```

This looks like red noise because the power density decreases at a rate of $\frac{1}{f^2}$ where $f$ is frequency.
